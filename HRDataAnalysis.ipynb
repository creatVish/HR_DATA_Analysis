```python
plt.figure(figsize=(15, 5))

age_bins = [20, 30, 40, 50, 60]
age_labels = ['20-30', '30-40', '40-50', '50-60']

# Create a new column 'AgeGroup' based on age bins
gen['AgeGroup'] = pd.cut(gen['Age'], bins=age_bins, labels=age_labels, right=False)

#
attrition_yes_df = gen[gen['Attrition'] == 'Yes']


# Cross-tabulation to get counts
cross_tab = pd.crosstab(index=attrition_yes_df['AgeGroup'], columns=gen['AgeGroup'])

cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Attrition by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

```

```python
plt.figure(figsize=(5, 5))

age_bins = [20, 30, 40, 50, 60]
age_labels = ['20-30', '30-40', '40-50', '50-60']

# Create a new column 'AgeGroup' based on age bins
gen['AgeGroup'] = pd.cut(gen['Age'], bins=age_bins, labels=age_labels, right=False)

#
attrition_yes_df = gen[gen['Attrition'] == 'Yes']


# Cross-tabulation to get counts
cross_tab = pd.crosstab(index=attrition_yes_df['Department'], columns=gen['Department'])

cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Attrition by Department')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
